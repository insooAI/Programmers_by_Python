```python
def solution(numbers, hand):
    left_key = [1,4,7,'*']
    right_key = [3,6,9,'#']
    neutral_key = [2,5,8,0]
    
    d = {'left':'L',
         'right' : 'R'}
    
    distance = {tuple(left_key) : 1,
                tuple(neutral_key) : 2,
                tuple(right_key) : 3}
    
    answer = ''
    
    location = {'L' : ['*'],
                'R' : ['#']}
    
    for number in numbers:
        # print(f'현재 눌러야할 숫자 : {number}')
        
        if number in left_key:
            # print('현재 왼손 엄지의 위치', location['L'][-1])
            # print('현재 오른손 엄지의 위치', location['R'][-1])
            answer += 'L'
            location['L'].append(number)

            # print('answer', answer)
            # print()
       
        elif number in right_key:
            # print('현재 왼손 엄지의 위치', location['L'][-1])
            # print('현재 오른손 엄지의 위치', location['R'][-1])            
            answer += 'R'
            location['R'].append(number)

            # print('answer', answer)
            # print()
            
        
        elif number in neutral_key:
            idx = neutral_key.index(number)
            
            # step 1
            if location['L'][-1] in left_key and location['R'][-1] in right_key: 
                if left_key.index(location['L'][-1]) == idx and right_key.index(location['R'][-1]) != idx:
                    # print('현재 왼손 엄지의 위치', location['L'][-1])
                    # print('현재 오른손 엄지의 위치', location['R'][-1])                        
                    answer += 'L'
                    location['L'].append(number)

                    # print('answer', answer)
                    # print()
                    
                elif right_key.index(location['R'][-1]) == idx and left_key.index(location['L'][-1]) != idx:
                    # print('현재 왼손 엄지의 위치', location['L'][-1])
                    # print('현재 오른손 엄지의 위치', location['R'][-1])                             
                    answer += 'R'
                    location['R'].append(number)

                    # print('answer', answer)
                    # print()
                    
                elif left_key.index(location['L'][-1]) == idx and right_key.index(location['R'][-1]) == idx:
                    # print('현재 왼손 엄지의 위치', location['L'][-1])
                    # print('현재 오른손 엄지의 위치', location['R'][-1])                               
                    answer += d[hand]
                    location[d[hand]].append(number)

                    # print('answer', answer)
                    # print()                 
                    
                elif left_key.index(location['L'][-1]) != idx and right_key.index(location['R'][-1]) != idx:
                    if left_key.index(location['L'][-1]) < right_key.index(location['R'][-1]):
                        # print('현재 왼손 엄지의 위치', location['L'][-1])
                        # print('현재 오른손 엄지의 위치', location['R'][-1])                             
                        answer += 'L'
                        location['L'].append(number)
                        
                    elif left_key.index(location['L'][-1]) > right_key.index(location['R'][-1]):
                        # print('현재 왼손 엄지의 위치', location['L'][-1])
                        # print('현재 오른손 엄지의 위치', location['R'][-1])                             
                        answer += 'R'
                        location['R'].append(number)                    
                        
                    elif left_key.index(location['L'][-1]) == right_key.index(location['R'][-1]):
                        # print('현재 왼손 엄지의 위치', location['L'][-1])
                        # print('현재 오른손 엄지의 위치', location['R'][-1])                             
                        answer += d[hand]
                        location[d[hand]].append(number)                     
                        
                        
            # step 2        
            # 왼쪽/오른쪽 엄지가 왼쪽/오른쪽 키패드에 없는 경우는, 중간 키패드에 있는 것
            
            elif location['L'][-1] not in left_key and location['R'][-1] in right_key: # 오른쪽에만 제대로 있는 경우
                dist_l = abs(neutral_key.index(location['L'][-1]) - idx)
                dist_r = abs(right_key.index(location['R'][-1]) - idx) + 1
                if dist_l > dist_r:
                    # print('현재 왼손 엄지의 위치', location['L'][-1])
                    # print('현재 오른손 엄지의 위치', location['R'][-1])           
                    answer += 'R'
                    location['R'].append(number)

                    # print('answer', answer)
                    # print()                   
                    
                elif dist_l < dist_r:
                    # print('현재 왼손 엄지의 위치', location['L'][-1])
                    # print('현재 오른손 엄지의 위치', location['R'][-1])                           
                    answer += 'L'
                    location['L'].append(number)

                    # print('answer', answer)
                    # print()                 
                    
                elif dist_l == dist_r:
                    # print('현재 왼손 엄지의 위치', location['L'][-1])
                    # print('현재 오른손 엄지의 위치', location['R'][-1])                           
                    answer += d[hand]
                    location[d[hand]].append(number)

                    # print('answer', answer)
                    # print()
                    
            elif location['L'][-1] in left_key and location['R'][-1] not in right_key: # 왼쪽에만 제대로 있는 경우
                dist_r = abs(neutral_key.index(location['R'][-1]) - idx)
                dist_l = abs(left_key.index(location['L'][-1]) - idx) + 1
                
                if dist_l > dist_r:
                    # print('현재 왼손 엄지의 위치', location['L'][-1])
                    # print('현재 오른손 엄지의 위치', location['R'][-1])                           
                    answer += 'R'
                    location['R'].append(number)

                    # print('answer', answer)
                    # print()                   
                    
                elif dist_l < dist_r:
                    # print('현재 왼손 엄지의 위치', location['L'][-1])
                    # print('현재 오른손 엄지의 위치', location['R'][-1])                           
                    answer += 'L'
                    location['L'].append(number)
                    # print('answer', answer)
                    # print()
                    
                elif dist_l == dist_r:
                    # print('현재 왼손 엄지의 위치', location['L'][-1])
                    # print('현재 오른손 엄지의 위치', location['R'][-1])                           
                    answer += d[hand]
                    location[d[hand]].append(number)

                    # print('answer', answer)
                    # print()
                    
            elif location['L'][-1] not in left_key and location['R'][-1] not in right_key: # 양쪽 다 없는 경우
                dist_l = abs(neutral_key.index(location['L'][-1]) - idx)
                dist_r = abs(neutral_key.index(location['R'][-1]) - idx)
                
                if dist_l > dist_r:
                    # print('현재 왼손 엄지의 위치', location['L'][-1])
                    # print('현재 오른손 엄지의 위치', location['R'][-1])                           
                    answer += 'R'
                    location['R'].append(number)

                    # print('answer', answer)
                    # print()
                    
                elif dist_l < dist_r:
                    # print('현재 왼손 엄지의 위치', location['L'][-1])
                    # print('현재 오른손 엄지의 위치', location['R'][-1])                          
                    answer += 'L'
                    location['L'].append(number)

                    # print('answer', answer)
                    # print()
                    
                elif dist_l == dist_r:
                    # print('현재 왼손 엄지의 위치', location['L'][-1])
                    # print('현재 오른손 엄지의 위치', location['R'][-1])                           
                    answer += d[hand]
                    location[d[hand]].append(number)

                    # print('answer', answer)
                    # print()
                    
    return answer
                    

                    
solution([1,4,4,6,7,9,5,2,3],'left')            


            
            
```

# 풀이 2
- map 함수 내 적용함수 f는 map에 의해 반복되므로 인덱싱할 필요 없음 주의!
- list indices must be integers or slices, not str 에러 : 리스트 인덱싱에 들어가는 것은 숫자 (문자열 불가)

```python
def solution(numbers, hand):
    
    matrix = [(4,2),(1,1),(1,2),(1,3),
              (2,1),(2,2),(2,3),
              (3,1),(3,2),(3,3),(4,1),(4,3)]
    
    answer = ''
    current_L = 10
    current_R = 11
    
    for number in numbers:
        if number in [1,4,7]:
            answer += 'L'
            current_L = number
            
        elif number in [3,6,9]:
            answer += 'R'
            current_R = number
            
        else:
            f = lambda x,y : abs(x-y)
            dist_L = sum(list(map(f, matrix[number], matrix[current_L])))
            dist_R = sum(list(map(f, matrix[number], matrix[current_R])))
            
            if dist_L < dist_R:
                answer += 'L'
                current_L = number
                
            elif dist_L > dist_R:
                answer += 'R'
                current_R = number
                
            else:
                if hand == 'left':
                    answer += 'L'
                    current_L = number
                    
                elif hand == 'right':
                    answer += 'R'
                    current_R = number
                    
    return answer
                    
                
                
            
    
```
